```python
import pandas as pd
import gspread
from gspread_dataframe import set_with_dataframe
```

```python
df = pd.read_csv('abalone.csv')
df = df[:50]
df
```

```python
GSHEET_NAME = 'Test'
TAB_NAME = 'Sheet1'

def write_df_to_sheet(GSHEET_NAME, TAB_NAME, df):
  gc = gspread.service_account(filename='test-gs-404002-61f72221cfb9.json')
  sh = gc.open(GSHEET_NAME)
  worksheet = sh.worksheet(TAB_NAME)
  set_with_dataframe(worksheet,df)
```

```python
write_df_to_sheet(GSHEET_NAME, TAB_NAME, df)
```
